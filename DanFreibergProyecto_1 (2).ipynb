```python
#Genero un modelo benchmark que relacione precio con superfice y cantidad de baños
X_benchmark=prop_ml[["surface_total","bathrooms"]]
y_benchmark=prop_ml[["price"]]
#Hago el train/test split 
from sklearn.model_selection import train_test_split
X_train_benchmark, X_test_benchmark, y_train_benchmark, y_test_benchmark =train_test_split(X_benchmark, y_benchmark, test_size=0.3,random_state=42)
```

```python
#Calculo los valores usados para entrenar el benchmark
len(y_train_benchmark)
```

```python
#Calculo los valores usados para testear el benchmark
len(y_test_benchmark)
```

```python
#Genero el arbol de decision a partir del benchmark y lo entreno
from sklearn.tree import DecisionTreeRegressor
arbol_benchmark=DecisionTreeRegressor(max_depth=3, random_state = 42)
arbol_benchmark.fit(X_test_benchmark,y_test_benchmark)
```

```python
#Calculo su efectividad
y_pred_benchmark=arbol_benchmark.predict(X_train_benchmark)
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_train_benchmark,y_pred_benchmark))
```

```python
#Hago mi arbol de decision con mis hiperparaemtros

X=prop_ml.drop(["price"],axis=1)
y=prop_ml[["price"]]
#Hago el train/test split 
from sklearn.model_selection import train_test_split
arbol=DecisionTreeRegressor(max_depth=5, random_state = 42)
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3,random_state=42)
arbol.fit(X_train,y_train)
```

```python
#Calculo su efectividad entrenandolo
y_pred_train=arbol.predict(X_train)
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_train,y_pred_train))
```

```python
#Calculo su efectividad con los valores test
y_pred_test=arbol.predict(X_test)
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred_test))
```

```python
#Hago un modelo de vecinos mas cercanos 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
vecino= KNeighborsRegressor(n_neighbors =5 , metric = 'minkowski', p = 2)
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3,random_state=42)
vecino.fit(X_train,y_train)
```
